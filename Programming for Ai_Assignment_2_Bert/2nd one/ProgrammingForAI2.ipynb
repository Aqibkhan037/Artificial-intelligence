```python
from transformers import pipeline
classifier = pipeline('sentiment-analysis')
```

```python
nltk.download("stopwords")
nltk.download('punkt')
stop_words= set(stopwords.words("english"))
```

```python
def remove_stopwods(df,label):
  j=0
  for i in df[label]:
    text_tokens=word_tokenize(i)
    filtered_words=[word for word in text_tokens if word not in stop_words]
    df[label][j]=filtered_words
    j=j+1
```

```python
df = pd.read_csv('/content/reviews_dataset.csv')
df = df.iloc[:764]
```

```python
df['news']= df['news'].astype("str")
```

```python
df['news'] = df['news'].apply(lambda x: x.lower())
df['news'] = df['news'].apply((lambda x: re.sub('[^a-zA-z0-9\s]','',x)))
```

```python
remove_stopwods(df,"news")
```

```python
def convertList(list1):
    str = ''  # initializing the empty string
    for i in list1: #Iterating and adding the list element to the str variable
      str =str+ " "+i
    return str
```

```python
k=0
while(k<len(df['news'])):
  df['news'][k]= convertList(df['news'][k])
  k=k+1
```

```python
templabel_BERT=[]
tempscores_BERT=[]
sentiment_statement=[]
for i in df['news']:
  templabel_BERT.append(classifier(i[:1000])[0]['label'])
  tempscores_BERT.append(classifier(i[:1000])[0]['score'])
  sentiment_statement.append(i)
```

```python
df["sentiment_label_bert"]= templabel_BERT
df["Sentiment_score"]= tempscores_BERT
```
